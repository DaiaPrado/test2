# Árboles de decisión: Clasificación

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import tree
from sklearn.datasets import make_classification
from sklearn import preprocessing
from sklearn.utils import shuffle
from sklearn.model_selection import train_test_split, KFold, GridSearchCV
from sklearn.datasets import load_wine
```

```python
def preproceso_datos(X,Y):
    X_reord, Y_reord = shuffle(X,Y,random_state=0)

    scaler = preprocessing.StandardScaler().fit(X_reord)
    X_escalada = scaler.transform(X_reord)

    X_train, X_test, Y_train, Y_test = train_test_split(X_escalada,Y_reord,train_size = 0.8, test_size = 0.2, random_state=0)
    return X_train, X_test, Y_train, Y_test
```

```python
def entrenamiento(X_train, Y_train):
    kf = KFold(shuffle=True, n_splits=7, random_state=0)
    model = tree.DecisionTreeClassifier()

    clf = model.fit(X_train,Y_train)
    
    #PLOT DEL ARBOL
    plt.figure(figsize=(20,10))
    tree.plot_tree(clf, max_depth = 4, impurity = True, fontsize=10)
    return clf
```

```python
# Todo: Entrena y representa el modelo de árbol de decisión
X, Y = make_classification(n_samples=1000, n_features=10, n_classes=2, shuffle=True, random_state=0)
X_train, X_test, Y_train, Y_test = preproceso_datos(X,Y)
best_model = entrenamiento(X_train, Y_train)
```
